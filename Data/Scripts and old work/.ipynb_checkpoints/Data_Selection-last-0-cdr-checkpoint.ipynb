```python
import pandas as pd
import numpy as np
```

```python
clinData = pd.read_csv('ADRC_Clinical_Data.csv')
```

```python
col_list = ['Label', 'Subject', 'cdr']
clinData = clinData[col_list]
clinData.loc[clinData.Label.apply(lambda x: x[-5]) == 'd', 'Days'] = clinData.Label.apply(lambda x: x[-4:])
clinData.loc[clinData.Label.apply(lambda x: x[-5]) != 'd', 'Days'] = clinData.Label.apply(lambda x: x[-5:])
# clinData['Days'] = clinData.Label.apply(lambda x: x[-4:])
clinData.Days = pd.to_numeric(clinData.Days, errors='coerce')
clinData = clinData.sort_values(by='Label', ascending=True)
clinData = clinData.reset_index(drop=True)
clinData
```

```python
cdr0  = clinData[clinData['cdr']==0]
subjects_cdr0 = cdr0['Subject'].unique()
len(subjects_cdr0)
```

```python
cdr1  = clinData[clinData['cdr']>0]
subjects_cdr1 = cdr1['Subject'].unique()
len(subjects_cdr1)
```

```python
df_start_0 = clinData.drop_duplicates(subset=['Subject'])
df_start_0 = df_start_0[df_start_0['cdr']==0]
df_start_0
```

```python
subjects_cdr_start0 = df_start_0['Subject'].unique()
len(subjects_cdr_start0)
```

```python
df_always_0 = clinData[clinData['Subject'].isin(subjects_cdr0)]
df_always_0 = df_always_0[~df_always_0['Subject'].isin(subjects_cdr1)]
df_always_0
```

```python
df_converters = clinData[clinData['Subject'].isin(subjects_cdr0)]
df_converters = df_converters[df_converters['Subject'].isin(subjects_cdr1)]
df_converters = df_converters.sort_values(by='Label', ascending=True)
df_converters = df_converters.reset_index(drop=True)
df_converters
```

```python
df_converted = df_converters[df_converters['cdr']>0]
df_converted = df_converted.drop_duplicates(subset=['Subject'])
df_converted
```

```python
df_c0 = df_converters[df_converters['cdr']==0]
df_c0
```

```python
df_joined = df_c0.merge(df_converted, on='Subject', suffixes=('','_converted'))
df_joined
```

```python
df_joined[df_joined['Days_converted']>500]
```

```python
hist = df_joined.hist(column='Days_converted', bins = 15)
```

```python
scans = pd.read_csv('Freesurfers.csv')
scans = scans[['Label']]
scans['Subject'] = scans.Label.apply(lambda x: x[:8])
# scans['Days'] = scans.Label.apply(lambda x: x[-4:])
scans.loc[scans.Label.apply(lambda x: x[-5]) == 'd', 'Days'] = scans.Label.apply(lambda x: x[-4:])
scans.loc[scans.Label.apply(lambda x: x[-5]) != 'd', 'Days'] = scans.Label.apply(lambda x: x[-5:])
scans.Days = pd.to_numeric(scans.Days, errors='coerce')
scans = scans.sort_values(by=['Label','Days'], ascending=True)
scans = scans.reset_index(drop=True)
scans = scans.drop_duplicates(subset=['Subject'])
# scans
```

```python
df_joined = df_joined.merge(scans, on='Subject', suffixes=('','_scan'))
df_joined
```

```python
df_joined['Days_scan_before_conversion'] = df_joined['Days_converted'] - df_joined['Days_scan']
df_joined['Days_scan_after_0'] = df_joined['Days_scan'] - df_joined['Days']
df_joined['Days_0_before_conversion'] = df_joined['Days_converted'] - df_joined['Days']
```

```python
df_joined
```

```python
df_joined = df_joined.sort_values(by='Days_0_before_conversion', ascending=True)
df_joined = df_joined[df_joined['Days_0_before_conversion']>0]
df_joined
```

```python
df_joined.sort_values(by='Days_scan_after_0', ascending=True)
df_joined = df_joined[df_joined['Days_scan_after_0']<365]
```

```python
df_joined
```

```python
# new df to only contain rows where Days_scan_after_0 < 365
# This means that the scan occurred less than 365 days after the patient was considered CN in their assessment
df_recent = df_joined[df_joined['Days_scan_after_0']<365]
# The number of days of the scan after 0 CDR should be less than the days before conversion
df_recent = df_recent[df_recent['Days_scan_after_0']< 0.5 * df_recent['Days_0_before_conversion']]
df_recent
```

```python
df_recent = df_recent.sort_values(by=['Label','Days_scan_after_0'], ascending=True)
df_recent = df_recent.reset_index(drop=True)
df_recent = df_recent.drop_duplicates(subset=['Subject'])
df_recent
```

```python
df_recent.to_csv('ChosenScans_converters.csv',index=False)
```
